```python
#1 Social Media – Post Engagement Analyzer

def post_engagement():
    n = int(input("Enter number of like entries: "))

    total_likes = 0

    for i in range(n):
        likes = int(input(f"Enter likes {i+1}: "))
        total_likes += likes

    print(f"Total Likes: {total_likes}")

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


post_engagement()
```

```python
#2 Healthcare – Medicine Stock Alert System

def medicine_stock():
  number_of_medicines = int(input("Enter number of medicines: "))
  medicine_stock_list = []
  for i in range(number_of_medicines):
    medicine_stock_list.append(int(input(f"Enter stock of medicine {i+1}: ")))

  for i in range(len(medicine_stock_list)):
     if medicine_stock_list[i] < 10:
        print(f"Low stock alert : medicine {i+1} is low in stock")

     else:
        print(f"Medicine {i+1} is in sufficient stock")

medicine_stock()
```

```python
#3 Agriculture – Rainfall Adequacy Checker

def rainfall_adequacy():
  number_of_days = int(input("Enter number of days: "))
  required_rainfall = float(input("Enter required rainfall: "))
  rainfall_list = []

  for i in range(number_of_days):
    rainfall_list.append(float(input(f"Enter rainfall for day {i+1}: ")))

  average_rainfall = sum(rainfall_list) / number_of_days

  print(f"Average rainfall for {number_of_days} days : {average_rainfall}")

  if average_rainfall >= required_rainfall:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: Inadequate Rainfall")

rainfall_adequacy()
```

```python
#4 Social Media – Duplicate Account Detection

def username():
   n = int(input("Enter number of People: "))
   people_list = []

   for i in range(n):
       people_list.append((input(f"Enter Username of person {i+1}: ")))

   people_set = set(people_list)

   if len(people_set) == len(people_list):
      print("Duplicate Accounts found : No ")

   else:
      print("Duplicate Accounts found : Yes")

username()
```

```python
#6 Agriculture – Premium Crop Price Filter

def filter_crops():
  n = int(input("Enter number of crops: "))

  premium_crops = []
  regular_crops = []

  for i in range(n):
    crop_price = int(input(f"Enter price of crop {i + 1}: "))

    if crop_price > 2000:
      premium_crops.append(crop_price)
    else:
      regular_crops.append(crop_price)

  print(f"Premium Crops: {premium_crops}")

filter_crops()
```

```python
#7 System Monitoring – Application Health Checker

def application_health_checker():
    error_count = int(input("Enter error count: "))

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")


application_health_checker()
```

```python
#8 Banking – Daily Transaction Limit Checker

def transaction_limit_checker():
    amount = int(input("Enter transaction amount: "))

    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")


transaction_limit_checker()
```

```python
#9 E-Learning – Student Attendance Eligibility System

def attendance_eligibility():
    days = int(input("Enter total number of days: "))

    attendance = []
    present_count = 0

    for i in range(days):
        status = input(f"Enter attendance for day {i+1} (P/A): ").lower()
        attendance.append(status)

        if status == "p":
            present_count += 1

    percentage = (present_count / days) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {eligibility}")


attendance_eligibility()
```

```python
#10 Smart Electricity Bill Analyzer
def electricity_bill_analyzer(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


units = int(input("Enter units consumed: "))
total_bill, status = electricity_bill_analyzer(units)

print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {status}")
```
