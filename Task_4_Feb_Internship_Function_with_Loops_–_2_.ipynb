```python
#1  Mobile Recharge Validation System

def mobile_recharge():
    plans = [199, 299, 399, 599]
    successful_recharge = False

    while successful_recharge == False:
        amount = int(input("Enter the amount for recharge : "))

        if amount < 50:
            print("Amount Invalid, enter valid amount")

        elif amount not in plans:
            print("Invalid Plan, enter right plan")

        else:
            successful_recharge = True
            print("Recharge Successful")

mobile_recharge()
```

```python
#2 Inventory Reorder Alert System

def products_stock():
  number = int(input("Enter number of products : "))
  products = {}

  for i in range(1, number + 1):
    name = input(f"Enter product {i} name : ")
    stock = int(input(f"Enter product {i} stock :"))
    products[name] = stock

  for keys, values in products.items():
    print(f"{keys} : {values}")

  all_good = True

  for key, value in products.items():
    if value < 15:
      all_good = False
      print(f"Reorder Alert : The stock of {key} is only {value}, Reorder more {15 - value} of {key} ")

  if all_good == True:
     print("All products in stock")

products_stock()
```

```python
#3 Student Result Processing System

def results():
  marks = []
  subjetcs = int(input("Enter number of subjects : "))

  for i in range(1, subjetcs + 1):
    marks.append(int(input(f"Enter marks of subject {i} : ")))

  avg = sum(marks) / subjetcs

  print(f"Marks : {marks}")
  print(f"Average marks : {avg}")

  if avg >= 50:
    print("Pass")

  else:
    print("Fail")

results()
```

```python
#4  Cab Fare Estimator with Retry Option

def cab_fare_calculator():
  retry = False

  while retry == False:
        base_fare = 50
        fare_per_km = 12
        distance = int(input("Enter distance amount in km : "))
        peak_hours = input("Is it peak hours (yes/no) : ").lower()

        if peak_hours == "yes":
           fare = (base_fare + (fare_per_km * distance)) * 0.25
           print(f"The total amount in peak hours is {fare}")

        elif peak_hours == "no":
           fare = base_fare + (fare_per_km * distance)
           print(f"The total amount in normal hours is {fare}")

        else:
           print("Invalid input, try again")
           continue

        user_retry = input("Do you want to retry (yes/no) : ").lower()

        if user_retry == "yes":
           retry = False

        elif user_retry == "no":
           retry = True

cab_fare_calculator()
```

```python
#5 Employee Attendance Eligibility
def attendence():

    days = int(input("Enter number of days : "))
    attendance = []

    for i in range(1, days + 1):

        while True:
            p_a = input(f"Enter attendance ( P / A ) of day {i} : ").lower()

            if p_a == "p" or p_a == "a":
                attendance.append(p_a)
                break

            else:
                print("Invalid input, please enter ( P / A )")

    print(f"Attendance : {attendance}")

    present_count = attendance.count("p")

    if present_count / days >= 0.75:
        return "Eligible"

    else:
        return "Not Eligible"


print(attendence())
```

```python
#6 Password Strength Checker

def password():
  flag = True

  while flag == True:
    password = input("Enter password : ")
    spl_char = ["@", "#", "$"]
    int_count = 0
    spl_count = 0
    satisfied_1 = True
    satisfied_2 = True
    satisfied_3 = True

    if len(password) < 8:
        print("Password should be atleast 8 characters")
        satisfied_1 = False

    for i in password:

        if i.isdigit():
            int_count += 1

        if i in spl_char:
            spl_count += 1

    if int_count < 1:
        print("Password should contain atleast one digit")
        satisfied_2 = False


    if spl_count < 1:
        print("Password should contain atleast one special character")
        satisfied_3 = False


    if satisfied_1 == True and satisfied_2 == True and satisfied_3 == True:
       print("Password accepted")
       print("Strong Password")
       flag = False

    else:
       print("Password not accepted")
       print("Weak Password")

password()
```
