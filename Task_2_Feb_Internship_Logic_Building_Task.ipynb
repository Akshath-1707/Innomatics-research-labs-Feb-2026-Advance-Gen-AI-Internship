```python
#1 Unique Words in a Sentence

sentence = input("Enter a sentence: ")

words = sentence.split()
unique_words = set(words)

print(f"Unique Words Count : {len(unique_words)}")
print(f"Unique Words : {unique_words}")
```

```python
#2 Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

maximum_salary = max(employees.values())
for key, values in employees.items():
    if values == maximum_salary:
        member = key

print(f"Highest Salary : {member} - {maximum_salary}")
```

```python
#3 Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
max = numbers[0]
min = numbers[0]
for i in range(1, len(numbers)):
  if max < numbers[i]:
    max = numbers[i]
  elif min > numbers[i]:
    min = numbers[i]

print(f"Maximum : {max}")
print(f"Minimum : {min}")
```

```python
#4 Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
  if i > 1000:
    count += 1

print(f"Products above 1000 : {count}")
```

```python
#5 Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
present_count = 0

for i in attendance:
  if i == "P":
    present_count += 1

present_percentage = (present_count / len(attendance)) * 100

print(f"Attendance Percentage : {present_percentage}%")
```

```python
#6 Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
phone_set = set(phone_numbers)

print(f"Unique phone numbers : {phone_set}")
```

```python
#7 Count Character Frequency
user_input = input("Enter a string: ")
char_count = {}

for i in user_input:
  if i in char_count:
    char_count[i] += 1
  else:
    char_count[i] = 1

print(char_count)
```

```python
#8 Convert List to Tuple
sample_list = [10,20,30]
sample_tuple = tuple(sample_list)
print(f"Tuple : {sample_tuple}")
```

```python
#9 Check if a Key Exists in a Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
user_input = input("Enter an employee name: ")

if user_input in employees:
  print("Employee Exits")
else:
  print("Employee does not exist")
```

```python
#10 Calculate Average Marks
marks = [45, 22, 89, 10, 66]
sum = 0

for i in marks:
  sum += i

average = sum / len(marks)
print(f"Average Marks : {average}")
```
