```python
#1 Smart Parking Lot Management System

def parking():
  parking_capacity = int(input("Enter the parking capacity: "))
  vehical_logs = list(input("Enter the vehical logs ( IN / OUT): ").lower().split())
  in_count = 0
  out_count = 0

  for i in vehical_logs:
    if i == "in":
      in_count += 1

    elif i == "out":
      out_count += 1

  current_count = in_count - out_count

  if current_count > parking_capacity:
    parking_status = "Not Available"
    print("Parking lot is full")
  else:
    parking_status = "Available"

  print(f"Currently parked vehicles: {current_count}")
  print(f"Parking status: {parking_status}")
parking()
```

```python
#2 Online Food Delivery Time Estimator

def food_delivery_time():

  distance = int(input("Enter the distance in km: "))
  traffic_level = input("Enter the traffic level (Low / Medium / High): ").lower()
  weather = input("Enter the weather condition (Sunny / Rainy): ").lower()

  avg_speed = 30

  time_taken = (distance / avg_speed) * 60

  if traffic_level == "low":
    time_taken *= 1
  elif traffic_level == "medium":
    time_taken *= 1.5
  elif traffic_level == "high":
    time_taken *= 2

  if weather == "sunny":
    time_taken += 0
  elif weather == "rainy":
    time_taken += 30

  print(f"Estimated delivery time: {time_taken} minutes")

food_delivery_time()
```

```python
#3 Movie Theatre Seat Occupancy Analyzer

def movie_theater_set():

  total_seats = int(input("Enter the total number of seats: "))
  booked_seats = list(input("Enter the number of booked seats: ").split())
  available_seats = total_seats - len(booked_seats)

  occupency_percentage = (available_seats / total_seats) * 100

  if occupency_percentage < 75:
     show_status = 'Available'
  elif occupency_percentage >= 75 and occupency_percentage < 100:
     show_status = 'Almsot Full'
  elif occupency_percentage == 100:
     show_status = 'Full'

  print(f"Occupency : {occupency_percentage} %")
  print(f"Show Status : {show_status}")

movie_theater_set()
```

```python
#4 Cloud Server Load Classification System

def cloud_server_load():
  cpu_readings = list(input("Enter the CPU readings (%): ").split())
  total = 0

  for i in cpu_readings:
    total += int(i)

  average = total / len(cpu_readings)

  if average < 50:
    server_status = 'Normal'

  elif average >= 50 and average < 80:
    server_status = 'Warning'

  else:
    server_status = 'Critical'

  print(f"Average CPU Load: {average} %")
  print(f"Server Status: {server_status}")

cloud_server_load()
```

```python
#5 Smart Classroom Resource Usage Monitor
def resource_usage():
  resources = {}
  n = int(input("Enter the number of resources: "))
  overused_resources = []

  for i in range(n):
    resource_name = input(f"Enter the resource name {i + 1}: ")
    usage = int(input(f"Enter the resource usage {i + 1}: "))
    resources[resource_name] = usage

  for keys, values in resources.items():
    if values > 6:
       overused_resources.append(keys)
       energy_alert = 'Yes'
    else:
       energy_alert = 'No'

  if len(overused_resources) == 0:
    print("No overused resources")

  else:
    print(f"Overused resources: {overused_resources}")

  print(f"Energy alert : {energy_alert}")
resource_usage()
```

```python
#6

def event_capacity():
  event_capacity = int(input("Enter the event capacity: "))
  registrations = int(input("Enter the number of registrations: "))

  if registrations > event_capacity:
     waitlisted_users = registrations - event_capacity
     status = 'Closed'
  else:
     waitlisted_users = 0
     status = 'Open'

  print(f"Confirmed registrations : {registrations}")
  print(f"Waitlisted users : {waitlisted_users}")
  print(f"Registration Status : {status}")

event_capacity()
```
