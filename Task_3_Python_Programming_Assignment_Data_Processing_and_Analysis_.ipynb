```python
#1 Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

maximum_bonus = max(employees.values())
emp_list = []

for keys, value in employees.items():
  if value == maximum_bonus:
    emp_list.append(keys)

print("Top Performers Eligible for Bonus:", ", ".join(emp_list),f"(Score = {maximum_bonus})")
```

```python
#2 Search Query Keyword Analysis
user_input = "Buy mobile phone buy phone online"
user_input_list = list(user_input.lower().split(" "))
user_input_dict = {}

for i in user_input_list:
  if i in user_input_dict:
    user_input_dict[i] += 1
  else:
    user_input_dict[i] = 1

user_input_dict = {keys: values for keys, values in user_input_dict.items() if values > 1}
print(user_input_dict)
```

```python
#3 Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
sensor_hour_readings = {}

for i in range(len(sensor_readings)):
  if sensor_readings[i] % 2 == 0:
     sensor_hour_readings[i] = sensor_readings[i]

sensor_hour_readings_list = []
for keys, values in sensor_hour_readings.items():
  sensor_hour_readings_list.append((keys, values))

print(sensor_hour_readings_list)
```

```python
#4 Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

gmail_count = 0
yahoo_count = 0

for i in emails:
   if i.endswith("gmail.com"):
      gmail_count += 1
   elif i.endswith("yahoo.com"):
      yahoo_count += 1

gmail_percentage = (gmail_count / len(emails)) * 100
yahoo_percentage = (yahoo_count / len(emails)) * 100

print(f"gmail.com = {gmail_percentage}%\n" f"yahoo.com = {yahoo_percentage}%")
```

```python
#5 Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
sales_avg = sum(sales) / len(sales)
sales_dict = {}
for i in range(len(sales)):
  if sales[i] > (sales_avg + (sales_avg * 0.30)):
    sales_dict[f"Day {i + 1} "] = sales[i]

print(sales_dict)
```

```python
#6 Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
user_id_dict = {}

for i in user_ids:
   if i in user_id_dict:
     user_id_dict[i] += 1
   else:
     user_id_dict[i] = 1

user_id_dict = {keys: values for keys, values in user_id_dict.items() if values > 1}
print(user_id_dict)
```
