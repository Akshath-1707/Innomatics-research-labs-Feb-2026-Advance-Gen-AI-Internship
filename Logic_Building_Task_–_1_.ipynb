```python
# question 1 : user login check

username = "admin"
password = "1234"

username_input = input("Enter username: ")
password_input = input("Enter password: ")

if username_input == username and password_input == password:
    print("Login successful")
else:
    print("Invalid Credentials")
```

```python
# question 2 : Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for i in marks:
  if i >= 50:
    pass_count += 1
  else:
    fail_count += 1

print("Total number of students passed : ", pass_count)
print("Total number of students failed : ", fail_count)
```

```python
# question 3 : Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

for i in range(len(names)):
    names[i] = names[i].strip()
    names[i] = names[i].lower()

print(names)
```

```python
# question 4 : Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
len_list = []

for i in range(len(messages)):
    len_count = len(messages[i])
    len_list.append(len_count)

    if len_count > 10:
        print(f"Flagged : {messages[i]}")
print(len_list)
```

```python
#question 5 : Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for i in logs:
    if i == "ERROR":
        error_count += 1

print("Total number of errors : ", error_count)
```
